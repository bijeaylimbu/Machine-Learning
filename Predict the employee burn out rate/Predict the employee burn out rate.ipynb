```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['Gender']=le.fit_transform(dataset['Gender'])
```

```
dataset['Company Type']=le.fit_transform(dataset['Company Type'])
```

```

dataset['WFH Setup Available']=le.fit_transform(dataset['WFH Setup Available'])
```

```
dataset
```

```
dataset['Date of Joining']=pd.to_datetime(dataset['Date of Joining'],format="%Y-%m-%d")
```

```
year=dataset['Date of Joining'].dt.year
```

```
dataset=dataset.assign(year=year.values)
```

```
import xgboost as xgb
```

```
xg_reg=xgb.XGBRegressor(objective="reg:linear",colsample_bytree=0.3,learning_rate=0.1,max_depth=5,alpha=10,n_estimators=10)
```

```
xg_reg.fit(dataset,dataset_label)
```
