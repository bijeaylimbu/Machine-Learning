### Training

```
import pandas as pd
```

```

dataset=pd.read_csv('Train.csv')
dataset
```

```
from sklearn.preprocessing import OneHotEncoder

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['Gender']=le.fit_transform(dataset['Gender'])
dataset['Relationship_Status']=le.fit_transform(dataset['Relationship_Status'])
dataset['Decision_skill_possess']=le.fit_transform(dataset['Decision_skill_possess'])
dataset['Compensation_and_Benefits']=le.fit_transform(dataset['Compensation_and_Benefits'])
dataset['Hometown']=le.fit_transform(dataset['Hometown'])
dataset
```

```
dataset.isnull().sum()
```

```
from numpy.random import randint
dataset['Age'].fillna(randint(19,65) ,inplace=True)
```

```
dataset['Unit'].drop
```

```
df=pd.DataFrame(dataset)
```

```
df.isnull().sum()
```

```
df['Time_of_service'].fillna(randint(0,43),inplace=True)
```

```
df.isnull().sum()
```

```
df.head()
```

```
df['Work_Life_balance'].value_counts()
```

```
df['Work_Life_balance'].fillna(df['Work_Life_balance'].mean(),inplace=True)
```

```
df.isnull().sum()
```

```
final_dataset=df.drop(['VAR2','VAR4'],axis=1)
```

```
final_dataset.isnull().sum()
```

```
final_data=final_dataset.drop(['Attrition_rate'],axis=1)
```

```
dataset_label=final_dataset['Attrition_rate'].copy()
```

```
final=final_data.drop('Employee_ID',axis=1)
```

```
dataset_label.head()
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(final,dataset_label)
```
